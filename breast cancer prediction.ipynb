Q: what is meaning of "Applies standardization to the training data." and what is a need for it
ans: 
1. The values in the dataset are converted into a common scale.
2. Each feature (column) is adjusted so that:
    i.Mean becomes 0
    ii.Standard deviation becomes 1
3. ex: Before Standardization = Age: 18, 25, 60
                             Salary: 20000, 50000, 100000
    After Standardization =  Age: -0.8, -0.2, 1.5
                             Salary: -1.1, 0, 1.4
4. The original information is still preserved, but the scale becomes balanced.
5. Here, Salary values are much larger than Age values.Many machine learning algorithms calculate:
    i.distance
    ii.gradients
    iii.weights
Because Salary numbers are huge, the model may think Salary is more important than Age even if it is not.


Algorithms That Need Standardization
Mostly important for:
1. KNN
2. SVM
3. Logistic Regression
4. Neural Networks
5. K-Means
6. PCA

Less important for:

1. Decision Trees
2. Random Forests
3. XGBoost

```python
sc = StandardScaler()  
```

```python
sc.fit(X_train)     #only training data to prevent any leakage
```

```python
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
#When you call .transform(), you are telling the computer to adjust the scale of your numerical features so they all fall within a similar range.
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
le = LogisticRegression()
```

```python
le.fit(X_train,y_train)                  #X = input, y = output
```

```python
y_pred = le.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
